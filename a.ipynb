```python
class Pila:
    ##constructor
    def __init__(self):
        self.capacidad = 20
        self.datos = [0]*self.capacidad
        self.size = 0

    def push(self, elemento):
        if(self.size<self.capacidad):
            self.datos[self.size] = elemento
            self.size+=1
        else:
            return -1

    def pop(self):
        if(self.size==0):
            print("Elementos")
            return None
        self.size -= 1
        return self.datos[self.size]

    def mostrarSiguiente(self):
        if(self.size==0):
            print("no hay elementos")
            return None
        return self.datos[self.size-1]
    def estaVacia(self):
        if self.size > 0: return self.size <=0

    def imp(self):
        resultado = "-"
        for i in range(self.size):
            resultado+=f"{self.datos[i]}"+"-"
        return resultado

    def __repr__(self):
        return self.imp()

    def match(self,element):
      mat = [')','(',']','[','}','{','>','<']
      return (mat[mat.index(element)+1]==self.datos[self.size-1])

pila = Pila()
listaindex = Pila()
s = list(input())
error = 0

print()
for i in range(len(s)):
  if s[i] == '(' or s[i] == '[' or s[i] == '{' or s[i] == '<':
    pila.push(s[i])
    listaindex.push(i)
  if s[i] == ')' or s[i] == ']' or s[i] == '}' or s[i] == '>':
    if(pila.match(s[i])):
      pila.pop()
    else:
      error = 1
      print("Error en", i)
      break

if error == 0:
  if pila.size == 0: print("Balanceada")
  else: print("Error en",listaindex.datos[0])
```
