```python
class Queue:
    def __init__(self):
        self.arr = [0] * 10
        self.start = 0
        self.end = 0
        self.size = 0
    
    def push(self, num):
        if self.size >= 10:
            return print(f"Error al agregar {num}. Lista llena")

        self.arr[self.end] = num
        self.end = (self.end + 1) % 10
        self.size += 1
        return
    
    def pull(self):
        if self.size <= 0:
            return print(f"Error al sacar. Lista vacia")
        
        x = self.arr[self.start]
        self.start = (self.start + 1) % 10
        self.size -= 1
        return x
    
    def showNext(self):
        return self.arr[self.start]
    
    def showContent(self):
        str = "["
        for i in range(0, self.size):
            str +=  f"{self.arr[(self.start + i) % 10]},"
        str += "]"

        print(f"Inicial: {self.start}")
        print(f"Final: {self.end}")
        print(f"Size: {self.size}")
        print(f"Estructura: {str}")

    def isFull(self):
        return self.size == 10
    
    def isEmpty(self):
        return self.size == 0

        
```

```python
q = Queue()

for i in range(0,12):
    q.push(i)

q.showContent()

for i in range(0, 12):
    q.pull()

q.showContent()

for i in range(0,8):
    q.push(i)

q.showContent()

for i in range(0,2):
    q.pull()

q.showContent()

for i in range(0,6):
    q.push(i)
    
q.showContent()
    
```
